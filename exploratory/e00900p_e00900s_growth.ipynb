# Growth of `e00900`, `e00900p` and `e00900s` in CPS

`e00900p` and `e00900s` should always sum to the same value, yet they do not after advancing to 2018.

```python
import taxcalc as tc
import pandas as pd
import numpy as np
```

```python
tc.__version__
```

```python
recs = tc.Records.cps_constructor()
calc = tc.Calculator(records=recs, policy=tc.Policy())
```

```python
df = calc.dataframe(['e00900p', 'e00900s', 'e00900'])
np.allclose(df.e00900, (df.e00900p + df.e00900s))
```

```python
calc.advance_to_year(2018)
```

```python
df = calc.dataframe(['e00900p', 'e00900s', 'e00900'])
np.allclose(df.e00900, (df.e00900p + df.e00900s))
```

```python
df['e00900_diff'] = df.e00900 - (df.e00900p + df.e00900s)
(df.e00900_diff.min(), df.e00900_diff.max())
```

```python
df.sort_values('e00900_diff')
```
