```python
import taxcalc as tc
import taxcalc_helpers as tch
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import seaborn as sns
```

```python
tch.set_plot_style()
```

```python
N = 10000
w = np.random.randint(100, size=N)
base = np.random.randint(100, size=N)
reform = base * 0.5 + 0.5 * base.sum() / N
```

```python
ax = tch.quantile_chg_plot(base, reform, w, w)
plt.title('Disposable income by decile, base and reform', loc='left'
plt.show()
```
